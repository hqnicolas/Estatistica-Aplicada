```python
# 7) Neste exercício, vamos explorar como manipular os índices de uma série.
# 7a) Renomeie os índices de uma série. Vamos supor que temos uma série chamada 
# 'serie_valores' com índices ['a', 'b', 'c', 'd'].

import pandas as pd

serie_valores = pd.Series([1, 2, 3, 4], index=['a', 'b', 'c', 'd'])
serie_valores.index = ['e', 'f', 'g', 'h']
print(f'Série com índices renomeados:')
print(f'{serie_valores}') 


# 7b) Reset os índices de uma série. Vamos supor que temos uma série chamada 
# 'serie_com_indices', sendo: serie_com_indices = pd.Series([10, 20, 30, 40], index=['a', 'b', 'c', 'd']).

serie_com_indices = pd.Series([10, 20, 30, 40], index=['a', 'b', 'c', 'd'])
serie_resetada = serie_com_indices.reset_index(drop=True)
print(f'Série com índices resetados:')
print(f'{serie_resetada}')


```
