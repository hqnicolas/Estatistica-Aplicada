```python
#vamos explorar como lidar com dados faltantes (NaN) em séries.
#8a) Crie uma série com valores ausentes (NaN), sendo: ([1, 2, np.nan, 4, np.nan]).
import pandas as pd
import numpy as np
ausentes = pd.Series([1, 2, np.nan, 4, np.nan])
print(f'Série com valores ausentes:')
print(f'{ausentes}')

#8b) Remova os valores ausentes da série anterior. 

ausentes = ausentes.dropna()
print(f'Série sem valores ausentes:')
print(f'{ausentes}')

#8c) Preencha os valores ausentes com o valor zero.

preenchidos = ausentes.fillna(0)
print(f'Série com valores ausentes preenchidos:')
print(f'{preenchidos}')
```
