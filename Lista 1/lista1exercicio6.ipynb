```python
# 6) Neste exercício, vamos explorar alguns métodos e funções úteis para manipular séries.
# 6a) Encontre os valores únicos em uma série. Vamos supor que temos uma série 
# chamada 'serie_duplicada', sendo: serie_duplicada = pd.Series([1, 2, 2, 3, 3, 4, 4]).

import pandas as pd

serie_duplicada = pd.Series([1, 2, 2, 3, 3, 4, 4])
valores_unicos = serie_duplicada.unique()
print(f'Valores únicos dentro desta série:')
print(f' {valores_unicos}')

# 6b) Verifique se um valor está presente em uma série. Vamos supor que temos uma 
# série chamada 'serie_numeros', sendo: serie_numeros = pd.Series([1, 2, 3, 4, 5]).

serie_numeros      = pd.Series([0, 1, 2, 3, 4, 5])
valor_existe    = serie_numeros.isin([1, 4])
print(f'Valores existentes na série:')
print(f'{valor_existe}')



# 6c) Ordene os elementos de uma série em ordem crescente. Vamos supor que temos 
# uma série chamada 'serie_desordenada', sendo: serie_desordenada = pd.Series([3, 1, 4, 2, 5]).

serie_desordenada = pd.Series([3, 1, 4, 2, 5])
serie_ordenada = serie_desordenada.sort_values(ascending=True)
print(f'Série ordenada em ordem crescente:')
print(f'{serie_ordenada}')
```
