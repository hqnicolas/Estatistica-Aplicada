```python
# 5) Neste exercício, vamos praticar algumas operações de manipulação de dados em séries.
# 5a) Adicione um novo elemento a uma série existente. Vamos supor que temos uma 
# série chamada 'serie_existente', sendo: serie_existente = pd.Series([1, 2, 3, 4], index=['a', 'b', 'c', 'd'])
import pandas as pd

serie_existente = pd.Series([1, 2, 3, 4], index=['a', 'b', 'c', 'd'])
nova_serie = pd.Series([5], index=['e'])
serie_adicionada = pd.concat([serie_existente, nova_serie])
print(f'Série adicionada:')
print(f'{serie_adicionada}')


# 5b) Remova um elemento de uma série. Vamos supor que temos uma série chamada 
# 'serie_valores', sendo: serie_valores = pd.Series([10, 20, 30, 40], index=['a', 'b', 'c', 'd']).

serie_valores  = pd.Series([10, 20, 30, 40], index=['a', 'b', 'c', 'd'])
serie_valores.drop('b', inplace=True)
print(f'Série depois da remoção do elemento "b":')
print(f'{serie_valores}')



# 5c) Altere o valor de um elemento específico em uma série.
# Vamos supor que temos uma série chamada 'serie_numeros', 
# sendo: serie_numeros = pd.Series([1, 2, 3, 4], index=['a', 'b', 'c', 'd']).

serie_numeros = pd.Series([1, 2, 3, 4], index=['a', 'b', 'c', 'd'])
serie_numeros['b'] = 5
print(f'Série com valor alterado:')
print(f'{serie_numeros}')
```
