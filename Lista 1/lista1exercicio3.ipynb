```python
# Crie uma série com os números de 1 a 20.
import pandas as pd
serie = pd.Series([1, 2, 3, 4, 5])
print(serie)

# 3a) Vamos praticar a indexação e seleção de elementos em uma série.
# Selecione o terceiro elemento da série criada no Exercício 01.

terceiro_elemento = serie[2]
print(f'Terceiro Elemento: {terceiro_elemento}')

# 3b) Selecione todos os elementos de uma série que sejam maiores que 10.
# Supondo que temos uma série chamada 'serie_valores' que contém os valores 
# ([5, 15, 8, 20, 10]).

serie = pd.Series([5, 15, 8, 20, 10])
maiores_que_10 = serie[serie > 10]
print(f' posição e Elemento da série maior que 10: ')
print(f'{maiores_que_10}')

# 3c) Selecione todos os elementos de uma série com índices entre 'b' e 'e'.
# Vamos supor que temos uma série chamada 'serie_letras', 
# sendo: serie_letras = pd.Series([1, 2, 3, 4], index=['a', 'b', 'c', 'd', 'e', 'f']).

serie_letras  = pd.Series([1, 2, 3, 4, 5, 6], index=['a', 'b', 'c', 'd', 'e', 'f'])
seleciona_bcde = serie_letras['b':'e']
print(f'Elementos da série com índices entre \'b\' e \'e\':')
print(f'{seleciona_bcde}')
```
