```python
# 4) Neste exercício, vamos realizar algumas operações básicas com séries.
# 4a) Some duas séries de números. Vamos supor que temos duas séries chamadas 
# 'serie1' e 'serie2'. Sendo elas, respectivamente, ([1, 2, 3, 4]) e ([5, 6, 7, 8]).

import pandas as pd
serie1 = pd.Series([1, 2, 3, 4])
serie2 = pd.Series([5, 6, 7, 8])
soma_series = serie1 + serie2
print(f'Soma das séries:')
print(f'{soma_series}')

# 4b) Multiplique uma série pelo número escalar 2. Vamos supor que temos uma série 
# chamada 'serie_numeros', sendo a série ([1, 2, 3, 4]). 

serie_numeros  = pd.Series([1, 2, 3, 4])
escalar  = serie_numeros * 2
print(f'Multiplicação da série por um número escalar:')
print(f'{escalar}')

# c) Calcule a média dos elementos de uma série. Vamos supor que temos uma série 
# chamada 'serie_valores', onde: serie_valores = pd.Series([10, 20, 30, 40])

serie_valores  = pd.Series([10, 20, 30, 40])
media_serie  = serie_valores.mean()
print(f'Média dos elementos da série:')
print(f'{media_serie}')
```
