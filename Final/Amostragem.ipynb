```python
import numpy as np
import pandas as pd
import random

datasetOriginal = pd.read_excel("Apostas Esportivas - Original.xlsx")

datasetOriginal.head()
```

```python
def amostragem_agrupamento(dataset, numero_grupos):
  intervalo = len(dataset) / numero_grupos

  grupos = []
  id_grupo = 0
  contagem = 0

  for _ in dataset.iterrows():
    grupos.append(id_grupo)
    contagem += 1

    if contagem > intervalo:
      contagem = 0
      id_grupo += 1

  print("Grupos criados: ", np.unique(grupos))

  dataset['grupo'] = grupos

  random.seed(1)
  grupo_selecionado = random.randint(0, numero_grupos - 1)

  print("Grupo selecionado: ", grupo_selecionado)
  print("Dataset shape: ", dataset.shape)

  return dataset[dataset['grupo'] == grupo_selecionado]
```

```python
datasetTratado = pd.read_excel("Apostas Esportivas - Tratado.xlsx")

datasetTratado.head()
```

```python
numero_grupos = 4

df_amostra_agrupamento = amostragem_agrupamento(datasetTratado, numero_grupos)
```

```python
print("Grupo selecionado shape: ", df_amostra_agrupamento.shape)
print("Quantidade elementos no grupo selecionado: ", df_amostra_agrupamento['grupo'].value_counts().values[0])
print("Frequência relativa do grupo: ", round((df_amostra_agrupamento['grupo'].value_counts().values[0] * 100)/len(dataset), 2))
```
